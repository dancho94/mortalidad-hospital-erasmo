```python
data = pd.get_dummies(data, columns=['SERVICIO'])
```

```python
data = data.drop('CODIGO', axis=1)
```

```python
#union de la columnas

data['SERVIVIO_PISO_5_12'] = data['SERVICIO_PISO 10'] | data['SERVICIO_PISO 11'] | data['SERVICIO_PISO 12'] | data['SERVICIO_PISO 5'] | data['SERVICIO_PISO 6'] | data['SERVICIO_PISO 7'] |data['SERVICIO_PISO 8'] | data['SERVICIO_PISO 9']
```

```python
data['SERVICIO_URGENCIA ADULTO'] = data['SERVICIO_URGENCIAS ADULTO'] | data['SERVICIO_URGENCIAS ADULTOS']
```

```python
data['SERVICIOS_URGENCIA PEDIATRIA'] = data['SERVICIO_URGENCIA PEDIATRIA'] | data['SERVICIO_URGENCIAS PEDIATRIA']
```

```python
#Elimina las columnas que utilize para unir

remove_urgencia = ['SERVICIO_URGENCIA PEDIATRIA', 'SERVICIO_URGENCIAS ADULTO', 'SERVICIO_URGENCIAS ADULTOS', 'SERVICIO_URGENCIAS PEDIATRIA']

data = data.drop(remove_urgencia, axis=1)
```

```python
remove_columns = ['SERVICIO_PISO 10', 'SERVICIO_PISO 11', 'SERVICIO_PISO 12', 'SERVICIO_PISO 5', 'SERVICIO_PISO 6',
                'SERVICIO_PISO 7', 'SERVICIO_PISO 8', 'SERVICIO_PISO 9']
data = data.drop(remove_columns, axis=1)
```

```python
data.head()
```

```python
data.dtypes
```

```python
data['FECHA'] = pd.to_datetime(data['FECHA'], format='%m/%d/%Y %H:%M:%S %p')
```

```python
data['FECHA NACIO'] = pd.to_datetime(data['FECHA NACIO'], format='%m/%d/%Y %H:%M:%S %p')
```

```python
data.dtypes
```

```python
replace_sexo = {'M' : '0', 'F' : '1'}

data['SEXO'] = data['SEXO'].replace(replace_sexo).astype(int)
```

```python
data.head()
```

```python
data['edad'] = (data['FECHA'] - data['FECHA NACIO']) // pd.Timedelta(days=365.25)
```
